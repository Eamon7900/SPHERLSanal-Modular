```python
# Input plot data here: ##########################################################################################
vars=["U0", "DM_r"] #The list of variables to be plotted against time.  Each will have its own figure:
zones=[5, 5, 7, 9, 11] #The list of zone numbers you would like to plot for each variable (The first zone doesn't plot for some reason, so just enter it twice):
timeTrim = 1 #How many time steps to trim from plot: 
plotPhase = False #Set to true to plot phase instead of time
##################################################################################################################

from matplotlib import pyplot as plt #Plot using pyplot
%run plots.py #A class to store colors, markers, and units for plots
%run DataParser.py
figCount = 1
for var in vars:
    figCount += 1
    plt.figure(figCount)
    plt.title(var + plots.units.get(var, ""))
    plt.ylabel(var + plots.units.get(var, ""))
    plt.xlabel("Time [s]")
    colors = plots.colors.copy()
    markers = plots.markers.copy()
    for zone in zones:
        varData = {
            "times" : [],
            "values" : []
        }
        timeCount = 0
        for modelDump in dataSet.data:
            timeCount += 1
            if timeCount <= timeTrim : continue

            if(plotPhase):
                modelDump.calcPhase(dataSet.period)
                varData["times"].append(modelDump.phase)
            else:
                varData["times"].append(modelDump.time)
            varData["values"].append(modelDump.grid[zone - 1][var])

        #If either the colors or the markers array is empty refill them
        if not colors: colors = plots.colors.copy()
        if not markers: markers = plots.markers.copy()
        #Plot the time series of the current zone
        print("plotting: " + str(varData["times"]) + str(varData["values"]))
        plt.plot(varData["times"], varData["values"], color=colors.pop(), marker=markers.pop(), label=str("Zone: " + str(zone)))


    
            

```
